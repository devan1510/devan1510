# parkinson disease (support vector machine)

```python
# imports for the dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the datset
df= pd.read_csv("C:\\classification problems\\parkinsons.csv")
df.head()
```

```python
df.drop("name",axis= 1,inplace= True)
```

# data preprocessing

```python
X= df.drop("status",axis= 1)
y= df.status
```

```python
# perform train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size= 0.2)
```

#  standarising data

```python
# turn the features to range btw 0 and 1
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
# training the model 
from sklearn.svm import SVC
model= SVC(kernel= "linear")
model.fit(X_train,y_train)
```

```python
# make prediction using the test dataset
predictions= model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
# make predictions on new inputs(feature values)
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569)
inp_array= np.asarray(input_data).reshape(1,-1)
```

```python
std_inp= scaler.transform(inp_array)
model.predict(std_inp)
```

# diabetes dataset(svm)

```python
# load the dataset
df= pd.read_csv("C:\\classification problems\\diabetes.csv")
df.head()
```

```python
df.drop_duplicates(inplace= True)
```

```python
# divide the dataset into train and test sets
X= df.drop("Outcome",axis= 1)
y= df.Outcome
```

```python
# perform test train split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size= 0.2)
```

```python
# perform feature scaling (min-max scaler)
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
model2= SVC(kernel= "rbf",)
```

```python
model2.fit(X_train,y_train)
```

```python
predictions= model2.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
