# principal component analysis(breast cancer)

```python
# imports for the projects
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# imports to upload the dataset
from sklearn.datasets import load_breast_cancer
```

```python
# load the dataset and convert into a pandas dataframe
cancer_data= load_breast_cancer()
df= pd.DataFrame(cancer_data.data,columns= cancer_data.feature_names)
df.head()
```

```python
# imports for scaling the data
from sklearn.preprocessing import StandardScaler
stand_scaler= StandardScaler()
scaled_data= stand_scaler.fit_transform(df)
```

```python
scaled_df= pd.DataFrame(scaled_data,columns= df.columns)
```

```python
# perform principal component analysis on the scaled_dtaa
from sklearn.decomposition import PCA
pca= PCA(n_components= 3)
pca_data= pca.fit_transform(scaled_data)
```

```python
scaled_df.shape,pca_data.shape
```

```python
pca_df= pd.DataFrame(pca_data)
```

```python
# visualise the data
fig= plt.figure(figsize= (10,10))
ax= fig.add_subplot(projection= "3d")
ax.scatter(data= pca_df,xs= "pca1",ys= "pca2",zs= "pca3",c=cancer_data.target)
```
